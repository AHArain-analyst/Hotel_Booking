## About Dataset
### Description
This data set contains booking information for a city hotel and a resort hotel, and includes information such as when the booking was made, length of stay, the number of adults, children, and/or babies, and the number of available parking spaces, among other things.

All personally identifying information has been removed from the data.

### Features
| Variable                                | Class    | Description                                                                                                           |
|-----------------------------------------|----------|-----------------------------------------------------------------------------------------------------------------------|
| hotel                                   | character | Hotel (H1 = Resort Hotel or H2 = City Hotel)                                                                      |
| is_canceled                             | double   | Value indicating if the booking was canceled (1) or not (0)                                                       |
| lead_time                               | double   | Number of days that elapsed between the entering date of the booking into the PMS and the arrival date            |
| arrival_date_year                       | double   | Year of arrival date                                                                                                 |
| arrival_date_month                      | character | Month of arrival date                                                                                                |
| arrival_date_week_number                | double   | Week number of year for arrival date                                                                               |
| arrival_date_day_of_month               | double   | Day of arrival date                                                                                                  |
| stays_in_weekend_nights                 | double   | Number of weekend nights (Saturday or Sunday) the guest stayed or booked to stay at the hotel                     |
| stays_in_week_nights                    | double   | Number of week nights (Monday to Friday) the guest stayed or booked to stay at the hotel                           |
| adults                                  | double   | Number of adults                                                                                                    |
| children                                | double   | Number of children                                                                                                  |
| babies                                  | double   | Number of babies                                                                                                    |
| meal                                    | character | Type of meal booked. Categories are presented in standard hospitality meal packages: Undefined/SC – no meal package; BB – Bed & Breakfast; HB – Half board (breakfast and one other meal – usually dinner); FB – Full board (breakfast, lunch and dinner) |
| country                                 | character | Country of origin. Categories are represented in the ISO 3155–3:2013 format                                          |
| market_segment                          | character | Market segment designation. In categories, the term "TA" means "Travel Agents" and "TO" means "Tour Operators"    |
| distribution_channel                    | character | Booking distribution channel. The term "TA" means "Travel Agents" and "TO" means "Tour Operators"                  |
| is_repeated_guest                       | double   | Value indicating if the booking name was from a repeated guest (1) or not (0)                                    |
| previous_cancellations                  | double   | Number of previous bookings that were cancelled by the customer prior to the current booking                      |
| previous_bookings_not_canceled           | double   | Number of previous bookings not cancelled by the customer prior to the current booking                            |
| reserved_room_type                      | character | Code of room type reserved. Code is presented instead of designation for anonymity reasons                           |
| assigned_room_type                      | character | Code for the type of room assigned to the booking. Sometimes the assigned room type differs from the reserved room type due to hotel operation reasons (e.g. overbooking) or by customer request. Code is presented instead of designation for anonymity reasons |
| booking_changes                         | double   | Number of changes/amendments made to the booking from the moment the booking was entered on the PMS until the moment of check-in or cancellation |
| deposit_type                            | character | Indication on if the customer made a deposit to guarantee the booking. This variable can assume three categories: No Deposit – no deposit was made; Non Refund – a deposit was made in the value of the total stay cost; Refundable – a deposit was made with a value under the total cost of stay. |
| agent                                   | character | ID of the travel agency that made the booking                                                                      |
| company                                 | character | ID of the company/entity that made the booking or responsible for paying the booking. ID is presented instead of designation for anonymity reasons |
| days_in_waiting_list                    | double   | Number of days the booking was in the waiting list before it was confirmed to the customer                           |
| customer_type                           | character | Type of booking, assuming one of four categories: Contract - when the booking has an allotment or other type of contract associated to it; Group – when the booking is associated to a group; Transient – when the booking is not part of a group or contract, and is not associated to other transient booking; Transient-party – when the booking is transient, but is associated to at least other transient booking |
| adr                                     | double   | Average Daily Rate as defined by dividing the sum of all lodging transactions by the total number of staying nights |
| required_car_parking_spaces             | double   | Number of car parking spaces required by the customer                                                            |
| total_of_special_requests                | double   | Number of special requests made by the customer (e.g. twin bed or high floor)                                     |
| reservation_status                       | character | Reservation last status, assuming one of three categories: Canceled – booking was canceled by the customer; Check-Out – customer has checked in but already departed; No-Show – customer did not check-in and did inform the hotel of the reason why |
| reservation_status_date                  | double   | Date at which the last status was set. This variable can be used in conjunction with the ReservationStatus to understand when was the booking canceled or when did the customer checked-out of the hotel |


### Data Source
The data is originally from the article [Hotel Booking Demand Datasets](https://www.sciencedirect.com/science/article/pii/S2352340918315191), written by Nuno Antonio, Ana Almeida, and Luis Nunes for Data in Brief, Volume 22, February 2019.

The data was downloaded and cleaned by [Thomas Mock and Antoine Bichat](https://github.com/rfordatascience/tidytuesday/blob/master/data/2020/2020-02-11/readme.md#data-dictionary) for #TidyTuesday during the week of February 11th, 2020.

## Purpose of Analysis

The dataset consists of various attributes related to hotel bookings, including information about guests, booking details, and hotel characteristics. The primary purpose of this analysis is to gain insights and understanding of the data to inform business decisions, improve hotel management, and enhance the guest experience. The key objectives of the analysis include:

1. **Booking Trends**: Explore trends in hotel bookings, such as the time of year when bookings are most frequent, the distribution of booking channels, and the average lead time between booking and arrival.

2. **Cancellation Analysis**: Investigate factors contributing to booking cancellations, including the impact of deposit type, market segment, and previous cancellations.

3. **Guest Demographics**: Understand the demographic profile of guests, including the number of adults, children, and babies, as well as the countries of origin for guests.

4. **Room Assignment**: Examine patterns in room assignments, including how often the assigned room type matches the reserved room type, and how booking changes affect room assignments.

5. **Pricing and Revenue**: Analyze the Average Daily Rate (ADR) to assess pricing strategies and revenue generation.

6. **Special Requests and Services**: Investigate the number of special requests made by guests and assess the impact of these requests on the guest experience.

7. **Customer Type Analysis**: Understand the different types of bookings, such as group bookings, contract bookings, transient, and transient-party, to tailor services accordingly.

8. **Parking and Other Amenities**: Evaluate the demand for car parking spaces and identify any additional services that can enhance guest satisfaction.

9. **Reservation Status**: Analyze the reservation status to identify patterns of cancellations, check-outs, and no-shows.

10. **Time Series Analysis**: Examine time series data to identify long-term trends and seasonal patterns, which can be used for revenue forecasting and resource allocation.

By conducting this analysis, we aim to provide actionable insights to the hotel management, marketing teams, and other stakeholders, allowing them to make informed decisions to optimize hotel operations, improve guest satisfaction, and maximize revenue.

### Data Analysis and Visualization

```python
# Calculate the normalized value counts of the 'is_canceled' column in the entire DataFrame 'df'
cancelled_percs = df['is_canceled'].value_counts(normalize=True)
# This provides the proportion of different 'is_canceled' values (canceled and not canceled) in the entire dataset

# Print the calculated normalized value counts
print(cancelled_percs)

# Create a new figure for the plot with a specific size
plt.figure(figsize=(5, 4))

# Set the title for the plot
plt.title('Reservation status count')

# Create a bar plot to show the count of 'is_canceled' values ('Not canceled' and 'Canceled')
# The x-axis labels are specified as ['Not canceled', 'Canceled']
# The heights of the bars are determined by the value counts of 'is_canceled' in the entire DataFrame
# The 'edgecolor' is set to 'k' (black) and 'width' is set to 0.7 for aesthetics
plt.bar(['Not canceled', 'Canceled'], df['is_canceled'].value_counts(), edgecolor='k', width=0.7)

# Show the plot
plt.show()
```

```python
# Create a new figure for the plot with a specific size
plt.figure(figsize=(8, 4))

# Create a countplot using Seaborn, showing the count of reservations for each hotel type
# The 'x' parameter specifies the 'hotel' column for the x-axis
# The 'hue' parameter distinguishes reservations based on the 'is_canceled' column (canceled or not canceled)
# The 'data' parameter specifies the DataFrame 'df' as the data source
# The 'palette' parameter sets the color palette for the plot
axl = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')

# Get the legend handles and labels from the countplot
legend_labels, _ = axl.get_legend_handles_labels()

# Set the title of the plot with a specific font size
plt.title('Reservation status in different hotels', size=20)

# Label the x-axis with 'hotel'
plt.xlabel('hotel')

# Label the y-axis with 'Number of reservations'
plt.ylabel('Number of reservations')

# Show the plot
plt.show()
```

```python
# Filter the DataFrame 'df' to select rows where the 'hotel' column is 'Resort Hotel'
resort_hotel = df[df['hotel'] == 'Resort Hotel']

# Calculate and display the normalized value counts for the 'is_canceled' column
# This provides the proportion of different 'is_canceled' values (canceled and not canceled) in the 'Resort Hotel' subset
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
# Filter the DataFrame 'df' to select only rows where the 'hotel' column is equal to 'City Hotel'
city_hotel = df[df['hotel'] == 'City Hotel']

# Calculate and display the normalized value counts of the 'is_canceled' column for the City Hotel
# This line of code counts the occurrences of each unique value in the 'is_canceled' column for City Hotel
# It normalizes the counts to show the proportions (frequencies as percentages) of each unique value
# It provides insight into the cancellation rate for City Hotel reservations
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
# Group the data in the 'resort_hotel' DataFrame by 'reservation_status_date'
# and calculate the mean of the 'adr' (Average Daily Rate) for each date
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()

# Group the data in the 'city_hotel' DataFrame by 'reservation_status_date'
# and calculate the mean of the 'adr' for each date
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
# Create a new figure for the plot with a specific size
plt.figure(figsize=(20, 8))

# Set the title for the plot with a large font size
plt.title('Average Daily Rate in City and Resort Hotel', fontsize=30)

# Plot the ADR (Average Daily Rate) for the Resort Hotel
# The x-values are taken from the index of the 'resort_hotel' DataFrame
# The y-values are taken from the 'adr' column of the 'resort_hotel' DataFrame
# A label 'Resort Hotel' is provided for this line in the legend
plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel')

# Plot the ADR for the City Hotel
# The x-values are taken from the index of the 'city_hotel' DataFrame
# The y-values are taken from the 'adr' column of the 'city_hotel' DataFrame
# A label 'City Hotel' is provided for this line in the legend
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel')

# Display the legend with a specified font size
plt.legend(fontsize=20)

# Show the plot
plt.show()
```

```python
# Extract the month from the 'reservation_status_date' column and create a new 'month' column.
df['month'] = df['reservation_status_date'].dt.month

# Create a figure for the plot with a specified size.
plt.figure(figsize=(16, 8))

# Create a countplot to show the number of reservations per month, differentiated by cancellation status.
axl = sns.countplot(x='month', hue='is_canceled', data=df)

# Get legend labels and handles for customization (commented out).
# legend_labels, _ = axl.get_legend_handles_labels()

# Set a title for the plot.
plt.title('Reservation status per month', size=20)

# Label the x-axis with 'month'.
plt.xlabel('month')

# Label the y-axis with 'number of reservations'.
plt.ylabel('number of reservations')

# Manually set the legend labels to 'not canceled' and 'canceled'.
plt.legend(['not canceled', 'canceled'])

# Display the plot.
plt.show()
```

```python
# Create a new figure with a specified size of 15x8 inches
plt.figure(figsize=(15, 8))

# Set the title for the plot with a large font size of 30
plt.title('ADR (Average Daily Rate) per month', fontsize=30)

# Filter the DataFrame to select only rows where reservations were canceled (is_canceled == 1).
# Then, group the data by the 'month' column, calculate the sum of 'adr' for each month, and reset the index.
canceled_df = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index()

# Create a bar plot (bar chart) using Seaborn, specifying the 'x' (month) and 'y' (adr) values.
# The 'data' parameter is set to the 'canceled_df' DataFrame.
sns.barplot(x='month', y='adr', data=canceled_df)

# Display the plot on the screen
plt.show()
```

```python
# Filter the DataFrame to select only the rows where 'is_canceled' is equal to 1 (indicating canceled reservations)
cancelled_data = df[df['is_canceled'] == 1]

# Count the number of canceled reservations for each unique country and select the top 10 countries
top_10_countries = cancelled_data['country'].value_counts()[:10]

# Create a new figure for the pie chart with a specific size
plt.figure(figsize=(8, 8))

# Set the title for the pie chart
plt.title('Top 10 countries with reservation canceled')

# Create a pie chart using the data from 'top_10_countries'
# 'autopct' specifies the format of the percentage labels on the chart
# 'labels' specifies the labels for each slice of the pie chart based on the country names
plt.pie(top_10_countries, autopct='%.2f', labels=top_10_countries.index)

# Display the pie chart
plt.show()
```

```python
# Calculate the average daily rate (ADR) for canceled reservations
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()

# Reset the index to make 'reservation_status_date' a regular column
cancelled_df_adr.reset_index(inplace=True)

# Sort the DataFrame by 'reservation_status_date' for better visualization
cancelled_df_adr.sort_values('reservation_status_date')

# Filter the DataFrame 'df' to select rows where 'is_canceled' is 0 (i.e., not canceled reservations)
not_cancelled_data = df[df['is_canceled'] == 0]

# Calculate the ADR for not canceled reservations
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()

# Reset the index to make 'reservation_status_date' a regular column
not_cancelled_df_adr.reset_index(inplace=True)

# Sort the DataFrame by 'reservation_status_date' for better visualization
not_cancelled_df_adr.sort_values('reservation_status_date')

# Create a new figure for the plot with a specific size
plt.figure(figsize=(20, 6))

# Set the title for the plot with a large font size
plt.title('Average daily rate', fontsize=30)

# Plot the ADR for not canceled reservations
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='not_cancelled')

# Plot the ADR for canceled reservations
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='cancelled')

# Display the legend to distinguish the lines in the plot
plt.legend()
```
